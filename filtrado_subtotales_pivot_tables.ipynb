```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'Región': ['Norte', 'Norte', 'Sur', 'Sur', 'Este', 'Este', 'Oeste', 'Oeste', 'Norte'],
    'Categoría': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A'],
    'Producto': ['X', 'Y', 'X', 'Y', 'Z', 'X', 'Y', 'Z', 'Z'],
    'Ventas': [100, 150, 200, 250, 300, 350, 400, 450, 500],
    'Gastos': [30, 45, 50, 60, 75, 85, 90, 100, 110]
}

df = pd.DataFrame(data)
df

```

```python
df_filtrado = df[df['Región'].isin(['Norte', 'Sur'])]
df_filtrado
```

```python
pivot_ventas = pd.pivot_table(df, values='Ventas', index='Región', columns='Categoría', aggfunc='sum')
pivot_ventas
```

```python
pivot_categoria_a = pivot_ventas.loc[:, 'A'];
pivot_categoria_a
```

```python
pivot_categoria_a = pivot_ventas.loc['Este'];
pivot_categoria_a
```

```python
pivot_totales = pivot_ventas.copy()
# Añadir fila de total general
pivot_totales.loc['Total'] = pivot_totales.sum()
# Añadir columna de total general
pivot_totales['Total'] = pivot_totales.sum(axis=1)


pivot_totales
```

```python
pivot_con_margins = pd.pivot_table(df, values ='Ventas', index=['Región', 'Categoría'], columns='Producto', aggfunc='sum', margins=True, margins_name='Total')
pivot_con_margins
```

```python
pivot_con_margins.fillna(0, inplace=True)
pivot_con_margins
```
