```python
import pandas as pd

data = {
    'Región': ['Norte', 'Norte', 'Sur', 'Sur', 'Este', 'Este', 'Oeste', 'Oeste', 'Norte'],
    'Categoría': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A'],
    'Ventas': [100, 150, 200, 250, 300, 350, 400, 450, 500],
    'Gastos': [30, 45, 50, 60, 75, 85, 90, 100, 110]
}

df = pd.DataFrame(data)
df
```

```python
pivot_ventas = pd.pivot_table(df, values='Ventas', index='Región', columns='Categoría', aggfunc='sum')
pivot_ventas
```

```python
pivot_promedio_ventas = pd.pivot_table(df, values='Ventas', index='Región', columns='Categoría', aggfunc='mean')
pivot_promedio_ventas
```

```python
pivot_ventas_gastos = pd.pivot_table(df, values=['Ventas', 'Gastos'], index='Región', columns='Categoría', aggfunc=['mean', 'sum'])
pivot_ventas_gastos
```

```python
pivot_ventas_gastos.fillna(0, inplace=True)
pivot_ventas_gastos
```

```python
pivot_ventas_gastos = pd.pivot_table(df, values=['Ventas', 'Gastos'], index='Región', columns='Categoría', aggfunc = {'Gastos':'mean', 'Ventas':'sum'})
pivot_ventas_gastos.columns = ['Ventas_Suma_A', 'Ventas_Suma_B', 'Gastos_Promedio_A', 'Gastos_Promedio_B']
pivot_ventas_gastos
```

```python
def rango(x):
    return x.max() - x.min()

pivot_rango = pd.pivot_table(df, values='Ventas', index='Región', columns='Categoría', aggfunc=rango)
pivot_rango
```
