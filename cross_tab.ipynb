```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'Región': ['Norte', 'Norte', 'Sur', 'Sur', 'Este', 'Este', 'Oeste', 'Oeste', 'Norte'],
    'Categoría': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A'],
    'Producto': ['X', 'Y', 'X', 'Y', 'Z', 'X', 'Y', 'Z', 'Z'],
    'Ventas': [100, 150, 200, 250, 300, 350, 400, 450, 500],
    'Gastos': [30, 45, 50, 60, 75, 85, 90, 100, 110]
}

df = pd.DataFrame(data)
df
```

```python
cross_tab = pd.crosstab(df['Región'], df['Categoría'])
cross_tab
```

```python
cross_tab = pd.crosstab([df['Región'], df['Categoría']], df['Producto'])
cross_tab
```

```python
normalize_table = pd.crosstab(df['Región'], df['Categoría'], normalize='index')
normalize_table
```

```python
normalize_column = pd.crosstab(df['Región'], df['Categoría'], normalize='columns')
normalize_column
```

```python
table_totals = cross_tab.copy()
table_totals.loc['Total'] = table_totals.sum(axis=0)
table_totals['Total'] = table_totals.sum(axis=1)
table_totals
```

```python
mean_seals = pd.crosstab(df['Región'], df['Categoría'], values=df['Ventas'], aggfunc='mean')
mean_seals
```

```python
def rango(x):
    return x.max() - x.min()
```

```python
range_seals = pd.crosstab(df['Región'], df['Categoría'], values=df['Ventas'], aggfunc=rango)
range_seals
```
